```python
fileArmiCategoria = pd.read_csv('Weapon category.csv')
fileArmiCategoria
```

```python
typeW = []
number = []
for i in range(len(fileArmiCategoria['Equipment_Type'])):
    for v in range(len(fileArmiCategoria[fileArmiCategoria.Equipment_Type == fileArmiCategoria['Equipment_Type'][i]].values[0][1:])):
        typeW.append(fileArmiCategoria['Equipment_Type'][i])
        number.append(fileArmiCategoria[fileArmiCategoria.Equipment_Type == fileArmiCategoria['Equipment_Type'][i]].values[0][v+1])

        
fileArmiCategoriaFormattato = pd.DataFrame({
    'Type':typeW,
    'Number':number
})
fileArmiCategoriaFormattato
```

```python
fig, ax = plt.subplots(figsize = (14,7))
sns.set_style('whitegrid')
#plt.grid()
plt.title('Distribuzione Tipi di Armi Vendute (1950-2020)',fontsize=20)
plt.xticks(fontsize=14,rotation=45)
plt.yticks(fontsize=12,rotation=45)
plt.xlabel('')
plt.ylabel('')
sns.boxplot( x=fileArmiCategoriaFormattato["Number"], y=fileArmiCategoriaFormattato["Type"],notch=True ,palette='deep')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (14,7))
plt.title('Distribuzione Tipi di Armi Vendute (1950-2020)',fontsize=20)
plt.xticks(fontsize=14,rotation=45)
plt.yticks(fontsize=12,rotation=45)
plt.xlabel('')
plt.ylabel('')
sns.violinplot( x=fileArmiCategoriaFormattato["Number"], y=fileArmiCategoriaFormattato["Type"],palette='deep')
plt.show()
```

```python
cinaExport = cina[cina.Tranfer_Type == 'Export']
russiaExport = russia[russia.Tranfer_Type == 'Export']
usaExport = usa[usa.Tranfer_Type == 'Export']
franciaExport = francia[francia.Tranfer_Type == 'Export']
germaniaExport = germania[germania.Tranfer_Type == 'Export']
```

```python
tot = cinaExport['2020'].values[0] + russiaExport['2020'].values[0] + usaExport['2020'].values[0] + franciaExport['2020'].values[0] + germaniaExport['2020'].values[0]
tot
```

```python
fig, ax = plt.subplots(figsize = (14,7))
names = ['Cina', 'Russia', 'USA', 'Francia','Germania','Altro']
size = [
    cinaExport['2020'].values[0],
    russiaExport['2020'].values[0],
    usaExport['2020'].values[0],
    franciaExport['2020'].values[0],
    germaniaExport['2020'].values[0],
    22800 - tot
]
 
my_circle = plt.Circle( (0,0), 0.7, color='white')

colors = sns.color_palette('deep')
plt.pie(size, labels=names, wedgeprops = {'edgecolor' : 'white' },colors=colors)
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```
