```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.patches as mpatches
```

```python
fileFameMondo = pd.read_csv('IndiceFameMondo.csv',sep=';')
fileFameMondo
```

```python
fig, ax= plt.subplots(figsize=(14,7))

plt.grid()
colors = sns.color_palette('deep')

plt.title('Indice di Carestia nel Mondo (Sbagliato)',fontsize=20)
plt.xticks(fontsize=14,rotation=45)
plt.xticks(fontsize=12,rotation=45)

for i in range(len(fileFameMondo)):
    val = pd.DataFrame({
      'x_axis': ['2000','2006','2012','2021'],
      'y_axis': fileFameMondo.loc[i][1:-2].values
    })
    plt.plot('x_axis', 'y_axis', data=val, linestyle='-',linewidth=2)
```

```python
fig, ax= plt.subplots(figsize=(14,7))

plt.grid()
colors = sns.color_palette('deep')

for i in range(len(fileFameMondo)):
    val = pd.DataFrame({
      'x_axis': ['2000','2006','2012','2021'],
      'y_axis': fileFameMondo.loc[i][1:-2].values
    })

    if fileFameMondo.loc[i]['Paese'] == 'Venezuela (Boliv. Rep. of)':
        plt.plot('x_axis', 'y_axis', data=val, linestyle='-',linewidth=3,color=colors[3],label='Venezuela')
    elif fileFameMondo.loc[i]['Paese'] == 'Angola':
        plt.plot('x_axis', 'y_axis', data=val, linestyle='-',linewidth=3,color=colors[2],label='Angola')
    else:
        plt.plot('x_axis', 'y_axis', data=val, linestyle='-',linewidth=2,color='grey',alpha=0.1,label='')
        
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)

plt.title('Indice di Carestia nel Mondo',fontsize=20)
plt.xticks(fontsize=14,rotation=45)
plt.xticks(fontsize=12,rotation=45)

plt.show()
```

```python
fileAfricaDenutrizione = pd.read_csv('AfricaMortalitaDenutrizione.csv',sep=';')
fileAfricaDenutrizione
```

```python
plt.figure(figsize=(10, 7))

plt.grid()

color = sns.color_palette("deep")
bar1 = sns.barplot(x="Zona",  y="Mortalita", data=fileAfricaDenutrizione, color=color[7],ci=None)
bar2 = sns.barplot(x="Zona", y='Denutrizione', data=fileAfricaDenutrizione, color=color[8],estimator=np.mean,ci=None)

mor = mpatches.Patch(color=color[7], label='Mortalità')
dem = mpatches.Patch(color=color[8], label='Denutrizione')

plt.title('Mortalità/Denutrizione in Africa',fontsize=20)
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=12, rotation=45)
plt.xlabel('')
plt.ylabel('')
plt.legend(handles=[mor, dem],fontsize=15)

plt.show()
```
